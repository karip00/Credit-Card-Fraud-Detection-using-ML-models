```python
import pandas as pd
import numpy as np
```

### Reading the Dataset and fetching all the necessary info

```python
data = pd.read_csv('creditcard.csv')
```

```python
pd.options.display.max_columns = None

data.head()
```

```python
print("Rows: ", data.shape[0])
print("Columns: ", data.shape[1])
```

```python
data.info()
```

### Dropping the Irrelevant Columns - Time

```python
data = data.drop(['Time'], axis=1)
```

```python
data.head()
```

### Checking null values

```python
data.isnull().sum()
```

### Handling Categorical Values - Binary Encoding

```python
data['Class'] = data['Class'].replace({'legit':0,'fraud':1})
```

```python
data.head()
```

### Feature Scaling - Amount

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
data['Amount'] = sc.fit_transform(data['Amount'].values.reshape(-1,1))
```

```python
data.head()
```

### Check Duplicates

```python
data.duplicated().any()
```

```python
data.shape
```

```python
data.drop_duplicates(inplace=True)
```

```python
data.shape
```

### Check if the dataset is balanced

```python
data['Class'].value_counts()
```

```python
import seaborn as sns

sns.countplot(data['Class'])
```

### Feature Matrix in X and Target Vector in y

```python
X = data.drop(['Class'], axis=1)
y = data['Class']
```

### Train - Test Split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=71)
```
